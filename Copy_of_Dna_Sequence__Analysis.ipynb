```python

#initialize empty string to store DNA sequence
sequence = ""

#read FASTA file and ignore header lines
with open("/content/brca1.fasta", "r") as file:
    for line in file:
        if not line.startswith(">"):
            sequence += line.strip().upper()

#calculate sequence length
length = len(sequence)

#count the number of nucleotides
A = sequence.count("A")
T = sequence.count("T")
G = sequence.count("G")
C = sequence.count("C")

#calculate the gc content
gc_content = ((G + C) / length) * 100

#print results
print("Sequence Length:", length)
print("A:", A)
print("T:", T)
print("G:", G)
print("C:", C)
print("GC Content (%):", gc_content)


```

## DNA Sequence Analysis using NCBI Data and Python
## Aim
to analyze the dna sequence using python and calculate GC content.
### Methods

1. The DNA sequence of the BRCA1 gene was retrieved from the NCBI nucleotide database in FASTA format.  
2. The FASTA file was uploaded and read using Python in Google Colab.  
3. Header lines starting with ">" were ignored, and the nucleotide sequence was extracted.  
4. Python was used to calculate the total sequence length and count the number of adenine (A), thymine (T), guanine (G), and cytosine (C) bases.  
5. GC content was calculated using the formula: (G + C) / total length Ã— 100.  
6. The results were printed and interpreted to understand nucleotide composition.
###### Results

The analyzed DNA sequence had a length of 48 nucleotides.  
The nucleotide composition showed higher GC content compared to AT content.  
The GC content of the sequence was calculated to be 75%.
