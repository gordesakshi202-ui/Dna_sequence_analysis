```python


sequence = ""

with open("/content/brca1.fasta", "r") as file:
    for line in file:
        if not line.startswith(">"):
            sequence += line.strip().upper()


length = len(sequence)


A = sequence.count("A")
T = sequence.count("T")
G = sequence.count("G")
C = sequence.count("C")

gc_content = ((G + C) / length) * 100


print("Sequence Length:", length)
print("A:", A)
print("T:", T)
print("G:", G)
print("C:", C)
print("GC Content (%):", gc_content)

#interpretation:-The DNA sequence was retrieved from the NCBI database and analyzed using Python. The GC content was calculated, which provides insight into the stability and structural properties of the gene. GC-rich regions are often associated with higher thermal stability and functional significance.
```
